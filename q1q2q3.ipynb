```python
ds = ds.fillna(ds.median())
```

```python
ds['horsepower'] = ds['horsepower'].astype('float64')  
ds.head()
```

```python
q1 = ds['mpg'].quantile(0.25)
q2 = ds['mpg'].median()
q3 = ds['mpg'].quantile(0.75)
iqr = q3 - q1

print("Q1:", q1)
print("Q2 (Median):", q2)
print("Q3:", q3)
print("IQR:", iqr)
```

```python
lo= q1 - 1.5 * iqr
up= q3 + 1.5 * iqr
lo
```

```python
up
```

```python
ds['mpg'] = ds['mpg'].apply(lambda x: lo if x < lo else x)
ds['mpg'] = ds['mpg'].apply(lambda x: up if x > up else x)
```

```python
print("\n'mpg' values after replacing outliers:")
print(ds['mpg'])
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
ds['mpg_minmax'] = scaler.fit_transform(ds[['mpg']])
scaler
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
ds['mpg_standard'] = scaler.fit_transform(ds[['mpg']])
scaler
```
